**Task 2 EDA

Step 1: Summary Statistics

```python
df.describe()
```

Step 2: Mean - Median - Mode - Standard Deviation

```python
df.mean(numeric_only=True)
df.median(numeric_only=True)
df.mode()
df.std(numeric_only=True)
```

Step 3: Histograms

```python
import matplotlib.pyplot as plt

df.hist(figsize=(18,15))
plt.tight_layout()
plt.show()
```

Step 4: Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Total day minutes'], df['Total day charge'])
plt.xlabel("Total day minutes")
plt.ylabel("Total day charge")
plt.title("Total Day Minutes vs Total Day Charge")
plt.show()
```

Step 5: Correlation

```python
corr = df.corr(numeric_only=True)
corr
```

```python
import seaborn as sns

plt.figure(figsize=(14,10))
sns.heatmap(corr,
            annot=True,
            fmt='.2f',
            cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

A perfect positive correlation (1.0) exists between call minutes and their corresponding charges because charges are directly calculated from minutes.
